```python
from matplotlib import pyplot as plt
from math import pi, cos, e
from cmath import phase
from numpy import linspace
```

# plotting X1(t)

```python
def func_x1(t):
    cos_arg = pi * t / 5e-6
    return (0.6+
        2 * 0.16 * cos(2 * cos_arg - 1.88) +
        2 * 0.26 * cos(4 * cos_arg - 0.627) +
        2 * 0.06 * cos(6 * cos_arg - 2.5) +
        2 * 0.037 * cos(8 * cos_arg + 1.84) +
        0.2 * cos(10 * cos_arg)
    )
```

```python
x = linspace(0, 5e-6, num=200)
y = [func_x1(t) for t in x]

plt.figure(figsize=(10, 3))
plt.plot(x, y, "b-")
plt.grid(color='black', linestyle='-', linewidth=1)
plt.show()
```

# Calculation of coefficients

```python
with open("output_data.txt", "w") as file:
    print(f"i\t\t\tx\t\t\t\t\t\tabs(x)\t\t\t\tArg(x)")
    print('-' * 120)
    for iterarion in range(1, 10):
        e_arg = -1j * 2 * pi * iterarion
        x = (1 / 20) * (
            e ** (e_arg * 0 / 20) +
            e ** (e_arg * 1 / 20) +
            e ** (e_arg * 2 / 20) +
            e ** (e_arg * 3 / 20) +
            e ** (e_arg * 4 / 20) +
            e ** (e_arg * 8 / 20) +
            e ** (e_arg * 9 / 20) +
            e ** (e_arg * 10 / 20) +
            e ** (e_arg * 11 / 20) +
            e ** (e_arg * 12 / 20)
            )
        x_abs = abs(x)
        x_phase = phase(x)
        print(f"{iterarion:15}, {x:15}, {x_abs:15}, {x_phase:15}")
        # print(f"{iterarion}\t{x}\t\t{abs(x)}\t\t{phase(x)}")
```

# Plotting X3(t)

```python
def func_x3(t):
    cos_arg = t * pi  / (5 * 10 ** (-6))
    return (0.5 + \
        2 * 0.13 * cos(2 * cos_arg - 1.88) +
        2 * 0.26 * cos(4 * cos_arg - 0.62) +
        2 * 0.12 * cos(6 * cos_arg - 2.5) +
        2 * 8.77e-17 * cos(8 * cos_arg + 0.33) +
        2 * 0.1 * cos(10 * cos_arg + 3.67e-16) +
        2 * 0.03 * cos(12 * cos_arg - 1.88) +
        2 * 0.06 * cos(14 * cos_arg - 0.62) +
        2 * 1.804e-16 * cos(16 * cos_arg - 1.03) +
        2 * 0.02 * cos(18 * cos_arg + 1.88) +
        0.1 * cos(20 * cos_arg)
    )
```

# Original spectrum 

```python
def func_spec_1(w):
    return abs(-(1j * (1 - e ** (-1j * (-6 + e) * w))) / w -
            (1j * e ** (-3 * 1j * (-2 + e) * w) * (-1 + e ** (1j * e * w))) / w)
```

```python
x_2 = linspace(-15, 15, num=500)
y_2 = [func_spec_1(t) for t in x_2]

plt.figure(figsize=(12, 3))
plt.plot(x_2, y_2, "b-")
plt.grid(color='black', linestyle='-', linewidth=1)
plt.show()
```
