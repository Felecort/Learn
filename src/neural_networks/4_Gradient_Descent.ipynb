# 3 Hot and Cold Learning

```python
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001

for iteration in range(1101):

    prediction = input * weight
    error = (prediction - goal_prediction) ** 2

    print(f"Error: {error:.4f} Prediction: {prediction:.4f}" )
    # print("Error:" + str(error) + " Prediction:" + str(prediction))
    
    up_prediction = input * (weight + step_amount)
    up_error = (goal_prediction - up_prediction) ** 2

    down_prediction = input * (weight - step_amount)
    down_error = (goal_prediction - down_prediction) ** 2

    if(down_error < up_error):
        weight = weight - step_amount
        
    if(down_error > up_error):
        weight = weight + step_amount
```

# 4 Gradient Descent light

```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight = weight - direction_and_amount

    print(f"Error: {error:.4f} Prediction: {pred:.4f}" )
```

```python
def neural_network(input, weight, goal_pred):
    alpha = 0.1
    for _ in range(100):
        pred = input * weight
        error = (pred - goal_pred) ** 2
        derivative = (pred - goal_pred) * input
        weight = weight - (derivative * alpha)
    return weight, pred, goal_pred, error   
  
```

```python
def neural_network(input, weight, goal_pred, alpha):
    for _ in range(20):
        pred = input * weight
        error = (pred - goal_pred) ** 2
        derivative = (pred - goal_pred) * input
        weight = weight - (derivative * alpha)
        print(f"error: {error:.4f}, pred: {pred:.4f}, derivative = {derivative:.4f}, weight: {weight:.4f}")
    return goal_pred, pred, weight, error
```
