```python
def relu(x):
    return (x > 0) * x # returns x if x > 0
                       # return 0 otherwise


def relu2deriv(output):
    return output>0 # returns 1 for input > 0
                    # return 0 otherwise
```

```python
def learn_neural_network(inputs, outputs, alpha, hidden_size, weights_0_1, weights_1_2):
    for _ in range(60):
        layer_2_err = 0
        for i in range(len(inputs)):
            layer_0 = inputs[i]
            layer_1 = np.dot(layer_0, weights_0_1)
            layer_1 = relu(layer_1)
            layer_2 = np.dot(layer_1, weights_1_2)
            
            layer_2_err += np.sum((layer_2 - outputs[i]) ** 2)
            
            layer_2_delta = layer_2 - outputs[i]
            layer_1_delta = layer_2_delta.dot(weights_1_2.T) * relu2deriv(layer_1)
            
            weights_1_2-= alpha * layer_1.T.dot(layer_2_delta)
            weights_0_1 -= alpha * layer_0.T.dot(layer_1_delta)
    return weights_0_1, weights_1_2

```
